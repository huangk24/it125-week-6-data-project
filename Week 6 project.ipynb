```python
import pandas as pd
```

# Dataset about NBA players in 2022 seassion

```python
db = pd.read_csv("latest_RAPTOR_by_player.csv")
db
```

```python
# the structural attributes of the dataset

print (db.shape)   # rows and columns
print (db.columns) # column names
print (db.dtypes)  # data types of each column
```

## Find the player(s) with the most minutes played in 2022 seasson:

```python
db[db.mp == db.mp.max()]
```

## Find the player(s) with the least minutes played in 2022 seasson:

```python
db[db.mp == db.mp.min()]
```

## Find the average minutes played by all NBA players in 2022 seasson:

```python
db.mp.mean()
```

## Five number summary for minutes played

```python
min_played = db[["mp"]]
min_played.describe()
```

## Find the player(s) with the highest raptor score in 2022 seasson:

```python
db[db.raptor_total == db.raptor_total.max()]
```

## Find the player(s) with the lowest raptor score in 2022 seasson:

```python
db[db.raptor_total == db.raptor_total.min()]
```

## Find the average raptor score of all players in 2022 seasson:

```python
db.raptor_total.mean()
```

## Five number summary for raptor score

```python
raptor_total = db[["raptor_total"]]
raptor_total.describe()
```

```python
import matplotlib.pyplot as plt  
```

## Draw scatter plot to see if minutes played has relationship with raptor score

```python
fig, ax = plt.subplots()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.scatter(db['mp'], db['raptor_total'])

plt.xlabel("Minutes played")
plt.ylabel("Raptor score (total)")
plt.xlim([db.mp.min() - 2, db.mp.max() + 2])
```

There is no clear relationship between minutes played and raptor score
